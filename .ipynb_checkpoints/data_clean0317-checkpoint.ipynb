```python
#Analysis of Time Series

import pandas as pd
```

```python
##Extract csvs into dataframes
conf_data = "csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv"
conf_df = pd.read_csv(conf_data)

death_data = "csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv"
death_df = pd.read_csv(death_data)

rec_data = "csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv"
rec_df = pd.read_csv(rec_data)
rec_df.head()
```

```python
col_labels=['Province/State','Country/Region','Lat','Long','Date']
newc_df=pd.melt(conf_df,id_vars=['Province/State','Country/Region','Lat','Long'])
newc_df.head()
```

```python
col_labels=['Province/State','Country/Region','Lat','Long','Date']
newd_df=pd.melt(death_df,id_vars=['Province/State','Country/Region','Lat','Long'])
newd_df.head()
```

```python
col_labels=['Province/State','Country/Region','Lat','Long','Date']
newr_df=pd.melt(rec_df,id_vars=['Province/State','Country/Region','Lat','Long'])
newr_df.head()
```

```python
newc_df['Status']='Confirmed'
newd_df['Status']='Deaths'
newr_df['Status']='Recovered'
```

```python
newr_df.head()
```

```python
merged_df = pd.concat([newc_df, newd_df, newr_df]) 
merged_df.head()
```
